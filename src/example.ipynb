```python
import torch
from transformers import AutoTokenizer

from example import visualize_layer_results
from logit_lens import (
    AutoModelForCausalLMWithAliases,
    LogitLens,
)
```

```python
# model_name = "meta-llama/Llama-3.2-1B"
model_name = "openai-community/gpt2"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLMWithAliases.from_pretrained(model_name)
```

```python
# Sample input text
input_text = "Once upon a time"

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt")

# Generate text
with torch.no_grad(), LogitLens(model, post_emb=True, post_layer=True, topn=5) as ll:
    model(**inputs)
    ll_result = ll.get_results()

fig = visualize_layer_results(
    result=ll_result[0],
    input_ids=inputs["input_ids"][0],
    tokenizer=tokenizer,
)
fig
```
